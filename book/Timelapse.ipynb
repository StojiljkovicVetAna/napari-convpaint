# Time-lapse

The plugin is designed to work not just with single images but also with stacks of images, in particular time-lapses. Here's an example with a fluorescence microscopy time-lapse of a cell:

<video width="620" controls 
        src=../images/convpaint_timelapse.webm 
        type=video/webm> 
</video>

````python
import napari
from napari_convpaint.conv_paint import ConvPaintWidget
from napari.utils import nbscreenshot
import matplotlib.pyplot as plt
````

````python
viewer = napari.Viewer()
self = ConvPaintWidget(viewer)
viewer.window.add_dock_widget(self);
````

````python
nbscreenshot(viewer)
````

By default if a layer has an additional dimension, it will be taken into account by the plugin. This means e.g. that multiple planes or time-points can be annotated. For example:

````python
im1 = viewer.screenshot(canvas_only=False)
````

````python
im2 = viewer.screenshot(canvas_only=False)
````

````python
fig, ax = plt.subplots(2,1, figsize=(20,10))
ax[0].imshow(im1)
ax[1].imshow(im2);
ax[0].set_axis_off()
ax[1].set_axis_off()
````

If we now run the training, *all annotations* will be taken into account. Once training is done, we can verify the quality of the segmentation on a single frame, pressing ```Predict single frame```.

````python
nbscreenshot(viewer)
````

 Once satisfied we can run ```Predict all frames``` to get the full segmentation:

````python
nbscreenshot(viewer)
````

## Ignoring dimensions

Note that in certain cases, the additional dimension doesn't represent time or Z but channels. In that case, one can force the plugin to ignore the dimension using the ```Mutlichannel image``` option.
